```python
data["Goals"].plot(kind='hist', bins = 20) # Goals
```

```python
# Dit is een histogram voor de aantal doelpunten die iedere speler heeft gemaakt.
# De distributie van deze histogram heeft een Lognormal Distribution.
# Wat ik uit de histogram kan halen, is dat de piek van de histogram in het begin een hoge y-waarde heeften het snel omhoog gaat. 
# Vanaf de piek daalt het snel, maar de daling wordt naarmate het eind minder steil.
# Daarom is de de distributie een Lognormaal.
```
