#### Voor deze opdracht, voer ik een bivarate analysis op de kolommen Goals + (shots) OnTarget uit.
Als hypothese, wil ik weten of het aantal schoten op doel invloed heeft op het aantal doelpunten die een speler maakt.
Ik ga ervan uit dat er een sterke relatie is tussen het aantal schoten die een speler maakt en het aantal doelpunten die het oplevert.

```python


resultGoalsOnTarget = data[['Goals', 'OnTarget']].corr()
```

```python
resultGoalsOnTarget.style.background_gradient(cmap='coolwarm', axis=None)
```

```python
data.plot(kind='scatter', x='Goals', y='OnTarget')
```

#### Uit deze analyse, kan ik volgens de pearson correlation concluderen 
dat er een linear verband is tussen het aantal schoten op doel(Shots on target) en het aantal doelpunten dat een speler maakt.
Met een score van 0,86 is de verband groot.
Als ik mijn hypothese vergelijk met de resultaten, dan komt het resultaat overeen met die van mijn hypothese.
