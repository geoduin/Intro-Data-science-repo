```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import chi2_contingency
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

from sklearn import tree
import graphviz
```

```python
from sklearn import tree
import graphviz

def plot_tree_regression(model, features):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree")

    # Display in the notebook
    return graph 

def calculate_rmse(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    
    return (((predictions - actuals) ** 2).sum() / len(actuals)) ** (1/2)

def regressionOfModel(features, testSet, trainingSet, target):
    print(features)
    dt_regression = DecisionTreeRegressor(max_depth = 3)
    dt_regression.fit(trainingSet[features], trainingSet[target])
    
    # Evaluating the trained model
    predictionsOnTrainset = dt_regression.predict(trainingSet[features])
    predictionsOnTestset = dt_regression.predict(testSet[features])
    
    rmseTrain = calculate_rmse(predictionsOnTrainset, trainingSet[target])
    rmseTest = calculate_rmse(predictionsOnTestset, testSet[target])

    print("RMSE on training set " + str(rmseTrain))
    print("RMSE on test set " + str(rmseTest))
    
    print()
```

```python
data = pd.read_csv('Data.csv')
data.loc[data["Country"] == 'France', "League"] = "France Ligue 1"
data.head()
```

## Goals

#### Split your dataset into a train (70%) and test (30%) set.

```python
training, test = train_test_split(data, test_size=0.3, random_state=32, stratify=data['Country']);
```

#### Use the train set to fit a DecisionTreeRegressor. You are free to to choose which columns you want to use as feature variables and you are also free to choose the max_depth of the tree.

```python
features = ['Matches_Played', 'Mins', 'Shots', 'OnTarget']
dt_regression = DecisionTreeRegressor(max_depth = 8)
dt_regression.fit(data[features], data['Goals'])
```

#### Use your decision tree model to make predictions for both the train and test set.

#### Calculate the RMSE for both the train set predictions and test set predictions.

```python
regressionOfModel(features, test, training, 'Goals')
```

#### Is the accurracy different? Did you expect this difference?

De verschillen tussen beide waarden zijn marginaal, ongeveer een verschil van 0.01. 
Ik had niet verwacht dat de verschillen tussen trainingsset en test set klein waren.

Ik kan ervan uitgaan dat de voorspellingen van doelpunten nauwkeurig is.

#### Use the plot_tree function above to create a plot of the decision tree. Take a few minutes to analyse the decision tree. Do you understand the tree?

```python
plot_tree_regression(dt_regression, features)
```
