```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import chi2_contingency
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

from sklearn import tree
import graphviz
```

```python
from sklearn import tree
import graphviz

def plot_tree_regression(model, features):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree")

    # Display in the notebook
    return graph 

def calculate_rmse(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    
    return (((predictions - actuals) ** 2).sum() / len(actuals)) ** (1/2)
```

```python
penguinTraining, penguinTest = train_test_split(penguins, test_size=0.3, random_state=32, stratify=penguins['species']);
```

```python
features = ['bill_length_mm']
```

```python
print(penguinTraining.shape)
print(penguinTest.shape)
```

```python
def regressionOfModel(features, testSet, trainingSet, target):
    print(features)
    dt_regression = DecisionTreeRegressor(max_depth = 3)
    dt_regression.fit(trainingSet[features], trainingSet[target])
    
    # Evaluating the trained model
    predictionsOnTrainset = dt_regression.predict(penguinTraining[features])
    predictionsOnTestset = dt_regression.predict(penguinTest[features])
    
    rmseTrain = calculate_rmse(predictionsOnTrainset, trainingSet[target])
    rmseTest = calculate_rmse(predictionsOnTestset, testSet[target])

    print("RMSE on training set " + str(rmseTrain))
    print("RMSE on test set " + str(rmseTest))
    
    print()
```

```python
dt_regression = DecisionTreeRegressor(max_depth = 3)
```

```python
dt_regression.fit(penguinTraining[features], penguinTraining['body_mass_g'])
```

```python
# Evaluating the trained model
predictionsOnTrainset = dt_regression.predict(penguinTraining[features])
predictionsOnTestset = dt_regression.predict(penguinTest[features])
```

```python
rmseTrain = calculate_rmse(predictionsOnTrainset, penguinTraining.body_mass_g)
rmseTest = calculate_rmse(predictionsOnTestset, penguinTest.body_mass_g)

print("RMSE on training set " + str(rmseTrain))
print("RMSE on test set " + str(rmseTest))
```

```python
features = ['bill_length_mm','flipper_length_mm']
regressionOfModel(features,penguinTest, penguinTraining,  'body_mass_g')
features = ['bill_depth_mm', 'flipper_length_mm']
regressionOfModel(features,penguinTest, penguinTraining,  'body_mass_g')
features = ['bill_depth_mm', 'bill_length_mm']
regressionOfModel(features,penguinTest, penguinTraining,  'body_mass_g')
features = ['bill_depth_mm', 'bill_length_mm', 'flipper_length_mm']
regressionOfModel(features,penguinTest, penguinTraining,  'body_mass_g')
features = ['bill_length_mm']
regressionOfModel(features,penguinTest, penguinTraining,  'body_mass_g')

features = ['bill_depth_mm']
regressionOfModel(features,penguinTest, penguinTraining,  'body_mass_g')

features = ['flipper_length_mm']
regressionOfModel(features,penguinTest, penguinTraining,  'body_mass_g')
```
