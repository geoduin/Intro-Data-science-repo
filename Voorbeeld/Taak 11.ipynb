```python
print('\n species')
groupedSet = data.groupby('species')
groupMean = groupedSet.mean()
print(groupMean)
```

```python
# Ik kan vanuit de resultaten afleiden dat er statistische verschillen zijn tussen grouped by species mean resultaten en de reguliere mean resultaten

# Bij flipper_length_mm en body_mass_g heeft Gentoo met afstand een grotere mean waarde met 217.19 en 5076.01, dan Adelie en Chinstrap.
# Bij bill_depth_mm heeft Chinstrap een marginaal grotere mean waarde dan Adelie met 18.42 tegen 18.34, maar een veel grotere waarde dan Gentoo.

# Bij bill_length_mm heeft Chinstrap een marginaal grotere mean waarde dan Adelie met 48.83 tegen 47.5, maar een veel grotere waarde dan Gentoo.
```

```python
sns.barplot(y="bill_length_mm", x="species", data=data)
```

```python
sns.barplot(y="bill_depth_mm", x="species", data=data)
```

```python
sns.barplot(y="flipper_length_mm", x="species", data=data)
```

```python
sns.barplot(y="body_mass_g", x="species", data=data)
```

```python
# Bij bill_length_mm is er overloop tussen Chinstrap en Gentoo, maar is er een groot verschill tussen die twee en Adelie.
# Bij bill_depth_mm, body_mass_g en flipper_length_mm is er een overlap tussen Chinstrap en Adelie, maar is er een groot verschil tussen Gentoo en de andere twee.
```

```python

print('\nIsland')
groupedSet = data.groupby('island')
groupMean = groupedSet.mean()
print(groupMean)
```
